```python
import numpy as np
import pandas as pd

import json

from urllib.request import  urlopen
from bs4 import BeautifulSoup
```

```python
# extracting the url
URL = 'https://en.wikipedia.org/wiki/List_of_Super_Friends_episodes'
```

```python
html = urlopen(URL)
soup = BeautifulSoup(html, 'xml')

table = soup.find_all('table',{'class': "wikitable plainrowheaders wikiepisodetable"})
```

```python
soup.find_all('a',{'class': "wikitable plainrowheaders wikiepisodetable"})
```

```python
for _ in table:
    print(_.prettify())
```

```python
# Scrapping the name of each episode
# strip('\"').strip('\"')
names = []
for tr in table:
    ep_name = tr.find_all('td', 'summary')
    for episode in ep_name:
        names.append(str(episode.text))
print(names[:4])
```

```python
# Scrapping the dates each Season was produced

dates = []
for tr in table:
    ep_date = tr.find_all('span')
    for date in ep_date:
        dates.append(date.text.strip().strip('(').strip(')').strip(''))

dates = sorted(list(set(filter(None,dates))))
print(dates[:4])
```

```python
# Scrapping the Description of each episode

descriptions = []
for tr in table:
    description = tr.find_all('td','description')
    for desc in description:
        descriptions.append(desc.text.strip('\n'))

descriptions[:4]
```

```python
# Testing the results

assert len(descriptions)==len(dates)
assert len(descriptions) == len(names)
assert len(set(descriptions)) == len(set(dates))
```

### The code passed all the test hence ready for writing either in json format or as a csv file

## Json Saving

```python
# Main Dictionary for the SuperFriends Scrapper

SuperFriends = {}
SuperFriends['Date'] = dates
SuperFriends['Name'] = names
SuperFriends['Description'] = descriptions
```

## Saving as a Pandas DataFrame

```python
df = pd.DataFrame(SuperFriends)
df['date_name'] = df['Date']+": "+df['Name']
df.drop(columns=['Date','Name'],inplace=True)
```

```python
# saving to a csv format
df.to_csv('SuperFriends.csv', index=0)
```
